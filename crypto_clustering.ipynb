```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import hvplot.pandas
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from pathlib import Path
```

```python
 # Use the following endpoint to fetch json data
url = "https://min-api.cryptocompare.com/data/all/coinlist"
```

```python
# Alternatively, use the provided csv file:
file_path = Path("../AWS/crypto_data.csv")
crypto_df_1 = pd.read_csv(file_path, index_col=0)
crypto_df_1.head(10)
```

```python
# Keep only cryptocurrencies that are trading

crypto_df_2 = crypto_df_1[crypto_df_1['IsTrading'] == True]
crypto_df_2.head(10)
```

```python
# Keep only cryptocurrencies with a working algorithm
crypto_df_3 = crypto_df_2[crypto_df_2["Algorithm"] != "N/A"]
crypto_df_3.head(10)
```

```python
# Remove the 'IsTrading' column
crypto_df_4 = crypto_df_3.drop(columns='IsTrading')
crypto_df_4.head(10)
```

```python
# Remove rows with at least 1 null value
crypto_df_5 = crypto_df_4.dropna()
crypto_df_5
```

```python
# Remove rows with cryptocurrencies having no coins mined
crypto_df_6 = crypto_df_5[crypto_df_5["TotalCoinsMined"]> 0]
crypto_df_6.head(10)
```

```python
# Drop rows where there are 'N/A' text values
crypto_df_7 = crypto_df_6[crypto_df_6!='N/A']
crypto_df_7.head(10)
```

```python
# Store the 'CoinName'column in its own DataFrame prior to dropping it from crypto_df 
coin_name = crypto_df_7["CoinName"]
coin_name
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm
crypto_df_8 = crypto_df_7.drop(columns='CoinName')
crypto_df_8.head(10)
```

```python
# Create dummy variables for text features
new_df = pd.get_dummies(crypto_df_8, columns=['Algorithm', 'ProofType'])
new_df.head()
```

```python
# Standardize data
new_df_scaled = StandardScaler().fit_transform(new_df)
new_df_scaled
```

```python
# Use PCA to reduce dimensions to 3 principal components
pca = PCA(n_components=3)

crypto_pca = pca.fit_transform(new_df_scaled)
```

```python
# Create a DataFrame with the principal components data
crypto_pca_df = pd.DataFrame(
    data=crypto_pca, columns=["principal component 1", "principal component 2", "principal component 3"], 
    index= crypto_df.index
)
crypto_pca_df.head()
```

```python
# Finding the Best Value for k Using the Elbow Curve
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_pca_df)
    inertia.append(km.inertia_)
    
# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
# Initialize the K-Means model
model = KMeans(n_clusters=4, random_state=0)
    
# Fit the model
model.fit(crypto_pca_df)

# Predict clusters
predictions = model.predict(crypto_pca_df)

# Create a new DataFrame including predicted clusters and cryptocurrencies features
concat_pca_df = pd.DataFrame({
    "CoinName": coin_name,
    "Algorithm": crypto_df.Algorithm,
    "ProofType": crypto_df.ProofType,
    "TotalCoinsMined": crypto_df.TotalCoinsMined,
    "TotalCoinSupply": crypto_df.TotalCoinSupply,
    "Principal Component 1": crypto_pca_df["principal component 1"],
    "Principal Component 2": crypto_pca_df["principal component 2"],
    "Principal Component 3": crypto_pca_df["principal component 3"],
    "Class": model.labels_,
    },index= crypto_df.index)
concat_pca_df.head(10)


```

```python
# Create a 3D-Scatter with the PCA data and the clusters
fig = px.scatter_3d(
    concat_pca_df,
    x="Principal Component 3",
    y="Principal Component 2",
    z="Principal Component 1",
    color="Class",
    symbol="Class",
    width=800,
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```
